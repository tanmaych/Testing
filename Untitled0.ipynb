```
# Install the libraries
import numpy as np
import pandas as pd 
import math
import pandas_datareader as web
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```
#Read data fom web
df = web.DataReader('TTM',data_source='yahoo',start='2012-01-01',end='2019-12-17')
df.head(5)
```

```
df.shape
```

```
plt.figure(figsize=(16,8))
plt.title('Tata Motors')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.plot(df['Close'])
plt.show()
```

```
df=df[['Close']]
df.head(4)
```

```
#Create a variable to predict 'x'  days into the future
future_days = 50
# create a new column 'x' days up
df['Prediction'] = df['Close'].shift(-future_days)
df.tail(4)
```

```
#Create the feature data set 'X' and convert it into a numpy array and remove the last 'x' rows/days
X = np.array(df.drop(['Prediction'], 1))[:-future_days]
print(X)
```

```
y = np.array(df['Prediction'])[:-future_days]
print(y)
```

```
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.25)
```

```
tree = DecisionTreeRegressor().fit(x_train,y_train)
lr= LinearRegression().fit(x_train,y_train)
```

```
x_f = df.drop(['Prediction'],1)[:-future_days]
x_f=x_f.tail(future_days)
x_f=np.array(x_f)
x_f
```

```
tree_p =tree.predict(x_f)
print(tree_p)
print()
lr_p=lr.predict(x_f)
print(lr_p)
```

```
pr =tree_p
valid = df[X.shape[0]:]
valid['Predictions']=pr
plt.figure(figsize=(16,8))
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.plot(df['Close'])
plt.plot(valid[['Close','Predictions']])
plt.legend(['Orig','Val','Pred'])
plt.show()
```

```
pr =lr_p
valid = df[X.shape[0]:]
valid['Predictions']=pr
plt.figure(figsize=(16,8))
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.plot(df['Close'])
plt.plot(valid[['Close','Predictions']])
plt.legend(['Orig','Val','Pred'])
plt.show()
```
